```python
from keras import backend as K
K.clear_session()
```

```python
import keras.models

weight_path = './climateChange_1.h5'
# Recreate the exact same model
new_model = keras.models.load_model(weight_path)
new_model.summary()
```

```python
new_model.layers.pop()
new_model = keras.models.Model(inputs=new_model.inputs, outputs=new_model.layers[-1].output)
new_model.summary()
```

```python
#compile

from keras.optimizers import Adam

opt = Adam(lr=0.001)
new_model.compile(optimizer=opt, loss=keras.losses.categorical_crossentropy, metrics=['accuracy'])
```

```python
#get_images

from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.vgg16 import preprocess_input

import numpy as np
import glob
import os

img_path = "/home/sogang03/school/ClimateChange_tsne/images"
images = list(glob.glob(os.path.join(img_path,"*/*.jpg" )))

images = [images[i] for i in range(1, len(images), 1)]

# analyze images and grab activations
activations = []
for idx,img_path in enumerate(images):
    print ("getting activations for %s %d/%d" % (img_path,idx,len(images)))
    nimage = load_img(images[idx], target_size=(224,224))
    nimage = img_to_array(nimage)
    nimage = nimage.reshape((1, nimage.shape[0], nimage.shape[1], nimage.shape[2]))
    nimage= preprocess_input(nimage)       
    activations.append(new_model.predict(nimage)[0])
```

```python
#TSNE
from sklearn.manifold import TSNE

X = np.array(activations)
tsne = TSNE(n_components=3, perplexity=30, verbose=2).fit_transform(X)
```

```python
# save file
x, y = tsne[:,0], tsne[:,1], tsne[:2]
x = (x-np.min(x)) / (np.max(x) - np.min(x))
y = (y-np.min(y)) / (np.max(y) - np.min(y))
```

```python
tsne_path = 'tsne_points.txt'
f = open(tsne_path, 'w')

for idx,img_path in enumerate(images):
	line = "%s, %f, %f" % (images[idx], x[idx], y[idx])
	f.write('%s\n'%line)

f.close()
```
